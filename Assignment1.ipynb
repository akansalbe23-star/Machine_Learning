### Q1

```python
#a Reverse the Numpy array
arr= np.array([1,2,3,6,4,5])
reversed_arr = arr[::-1]
print(reversed_arr)
```

```python
#b Flatten the array
array1 = np.array([[1,2,3],[2,4,5],[1,2,3]])
flat1 = array1.flatten()
print("original array")
print(array1)
#Method 2
flat2 = array1.ravel()
print("Flaten array :")
print("1st Method : ",flat1)
print("2st Method : ",flat2)
```

```python
#c Compare two arrays
arr1 = np.array([[1,2],[3,4]])
arr2 = np.array([[1,2],[3,4]])
print("Comparison between two arrays : ",np.array_equal(arr1,arr2))
```

```python
#d Most Frequent and indices
x = np.array([1,2,3,4,5,1,2,1,1,1])
y = np.array([1,1,1,2,3,4,2,4,3,3])

# For x
val_x = np.bincount(x).argmax()
indices_x = np.where(x == val_x)[0]

# For y
val_y = np.bincount(y).argmax()
indices_y = np.where(y == val_y)[0]

print(f"Most Freqent value for x : {val_x}\n indices for x : {indices_x}")
print(f"Most Freqent value for y : {val_y}\n indices for y : {indices_y}")
```

```python
#e Matrix sums
gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]')

total_sum = np.sum(gfg)
row_sum = np.sum(gfg, axis=1)
col_sum = np.sum(gfg, axis=0)

print("Total sum : \n",total_sum)
print("Row sum : \n",row_sum)
print("Col sum : \n",col_sum)
```

```python
#f matrix operations
n_array = np.array([[55, 25, 15],[30, 44, 2],[11, 45, 77]])

# i. Sum of diagonal
diag_sum = np.trace(n_array)

# ii. Eigen values & vectors
eigen_vals, eigen_vecs = np.linalg.eig(n_array)

# iii. Inverse
inverse = np.linalg.inv(n_array)

# iv. Determinant
det = np.linalg.det(n_array)

print("Diagonal sum : \n",diag_sum)
print("Eign Values : \n",eigen_vals)
print("Eigen Vector : \n",eigen_vecs)
print("Inverse Matrix : \n",inverse)
print("Determinant : \n",det)
```

```python
#(g) Matrix multiplication & covariance
# First case
p1 = np.array([[1,2],[2,3]])
q1 = np.array([[4,5],[6,7]])

mul1 = np.dot(p1,q1)

# Flatten both for covariance
cov1 = np.cov(p1.flatten(), q1.flatten())

# Second case
p2 = np.array([[1,2],[2,3],[4,5]])
q2 = np.array([[4,5,1],[6,7,2]])

mul2 = np.dot(p2,q2)

cov2 = np.cov(p2.flatten(), q2.flatten())

print("Matrix multiplication (1):\n", mul1)
print("Covariance (1):\n", cov1)

print("\nMatrix multiplication (2):\n", mul2)
print("Covariance (2):\n", cov2)

```

```python
# (h) Inner, outer, cartesian product
x = np.array([[2,3,4],[3,2,9]])
y = np.array([[1,5,0],[5,10,3]])

inner = np.inner(x, y)
outer = np.outer(x, y)
cartesian = np.cross(x, y)

print("Inner Product : \n",inner)
print("outer product : \n", outer)
print("Cartestian product : \n",cartesian)
```

### Q2: Mathematics & Statistics

```python
# (a) Array calculations
array = np.array([[1,-2,3],[-4,5,-6]])

# i. Absolute
abs_val = np.abs(array)
print("Absolute Value : \n",abs_val)
# ii. Percentiles
flat = array.flatten()
percentiles_flat = np.percentile(flat, [25,50,75])
percentiles_col = np.percentile(array, [25,50,75], axis=0)
percentiles_row = np.percentile(array, [25,50,75], axis=1)
print("percentile_flat : \n",flat)
print("percentiil_col : \n",percentiles_col)
print("percentile row : \n",percentiles_row)
# iii. Mean, Median, Std
mean_flat, median_flat, std_flat = np.mean(flat), np.median(flat), np.std(flat)
mean_col, median_col, std_col = np.mean(array,axis=0), np.median(array,axis=0), np.std(array,axis=0)
mean_row, median_row, std_row = np.mean(array,axis=1), np.median(array,axis=1), np.std(array,axis=1)
print("Mean_Flat : \n",mean_flat)
print("Median_Flat : \n",median_flat)
print("Standar_Deviation_Flat : \n",std_flat)
print("Mean_col : \n",mean_col)
print("Median_col : \n",median_col)
print("Standar_Deviation_col : \n",std_col)
print("Mean_row : \n",mean_row)
print("Median_row : \n",median_row)
print("Standar_Deviation_row : \n",std_row)
```

```python
# (b) Floor, Ceil, Truncate, Round
a = np.array([-1.8, -1.6, -0.5, 0.5,1.6,1.8,3.0])

floor = np.floor(a)
ceil = np.ceil(a)
trunc = np.trunc(a)
round_val = np.round(a)

print("Floor : \n",floor)
print("ceil : \n", ceil)
print("trunc : \n", trunc)
print("round : \n", round_val)
```

### Q3: Searching & Sorting

```python
# (a) Sorting
array = np.array([10,52,62,16,16,54,453])

sorted_arr = np.sort(array)
indices = np.argsort(array)
smallest4 = np.sort(array)[:4]
largest5 = np.sort(array)[-5:]

print("sorted array : \n",sorted_arr)
print("indices : \n", indices)
print("smallest4 : \n", smallest4)
print("largest 5 : \n", largest5)
```

```python
# (b) Integers and float values only
array = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])

integers = array[np.floor(array) == array]
floats = array[np.floor(array) != array]

print("intgers : \n",integers)
print("float : \n", floats)
```

### Q4: Image Handling

```python
# (a) Function to save image as array
from PIL import Image

def img_to_array(path):
    img = Image.open(path)
    arr = np.array(img)
    if len(arr.shape) == 2:  # Grayscale
        np.savetxt("image_gray.txt", arr, fmt="%d")
    else:  # RGB
        arr_reshaped = arr.reshape(-1, arr.shape[-1])
        np.savetxt("image_rgb.txt", arr_reshaped, fmt="%d")
    return "Saved Successfully"
```
